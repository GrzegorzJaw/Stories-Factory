```python
from dotenv import dotenv_values
from IPython.display import Markdown
from openai import OpenAI
```

```python
env = dotenv_values(".env")

openai_client = OpenAI(api_key=env["OPENAI_API_KEY"])
```

## Jak używać modelu `gpt-4o` do generowania tekstu?

```python
res = openai_client.chat.completions.create(
    model="gpt-4o",
    temperature=0,
    messages=[
        {
            "role": "user",
            "content": """
Napisz rozdział do opowiadania pt. 'Przygody Maćka w Świecie Sztucznej Inteligencji'.

Opowiadanie ma następujące podsumowanie:
Maciek postanawia wejść do fascynującego świata sztucznej inteligencji.
W trakcie swojej podróży spotyka różne postacie, które pomagają mu zdobyć wiedzę na temat AI.
Dzięki ich wsparciu, Maciek staje się mistrzem w tej dziedzinie i zaczyna pomagać innym w nauce.
Na końcu, wspólnie z innymi, buduje sztuczną inteligencję, która ułatwia ludziom codzienne życie.

Rozdziały:
- Pierwsze kroki i spotkania z mentorami
- Zdobywanie wiedzy i pierwsze sukcesy
- Stawanie się mistrzem AI
- Pomaganie innym w nauce
- Budowanie AI, która pomaga ludziom

Napisz treść rozdziału 'Pierwsze kroki i spotkania z mentorami'.
            """,
        },
    ],
)
```

```python
Markdown(res.choices[0].message.content)
```

## Zamykamy model w funkcji

```python
def generate_text(prompt):
    res = openai_client.chat.completions.create(
        model="gpt-4o",
        temperature=0,
        messages=[
            {
                "role": "user",
                "content": prompt,
            },
        ],
    )

    return res.choices[0].message.content
```

```python
generate_text("Wymyśl tytuł dla opowiadania o przyg odach Maćka w świecie sztucznej inteligencji.")
```

## Pomysły `gpt-4o` na użycie modelu

### 1. Automatyczne Generowanie Artykułów Blogowych
Model AI może być używany do generowania artykułów na bloga na dowolny temat. Wystarczy podać temat i kilka kluczowych punktów, a model stworzy pełny artykuł, który może być gotowy do publikacji po drobnych poprawkach.

### 2. Tworzenie Podsumowań Długich Dokumentów
Model może analizować długie dokumenty, raporty lub artykuły i tworzyć ich zwięzłe podsumowania. Jest to szczególnie przydatne dla osób, które potrzebują szybko przyswoić najważniejsze informacje z obszernych materiałów.

### 3. Generowanie Dialogów dla Chatbotów
Można używać modelu AI do generowania realistycznych dialogów dla chatbotów. Model może być trenowany na danych z interakcji z klientami, aby dostarczać odpowiedzi na pytania, rozwiązywać problemy i prowadzić naturalne rozmowy.

### 4. Kreatywne Pisanie (Storytelling)
AI może wspierać pisarzy i twórców w procesie kreatywnego pisania, generując pomysły na fabułę, rozwijając postacie czy tworząc dialogi. Może to być przydatne zarówno w literaturze, jak i w scenariuszach filmowych.

### 5. Automatyczne Tworzenie Pytań Egzaminacyjnych
Model AI może być używany do generowania pytań egzaminacyjnych na podstawie dostarczonych materiałów edukacyjnych. Pytania mogą być różnego typu – wielokrotnego wyboru, otwarte, prawda/fałsz.

### 6. Personalizowane Listy Zakupów
Na podstawie preferencji użytkownika, historii zakupów i prognoz dotyczących zużycia, AI może generować spersonalizowane listy zakupów, które będą odpowiadały potrzebom użytkownika na dany okres.

### 7. Automatyczne Tworzenie Treści Marketingowych
Model AI może generować treści marketingowe, takie jak opisy produktów, e-maile reklamowe czy posty w mediach społecznościowych, dostosowane do konkretnej grupy docelowej i strategii marki.

### 8. Ulepszanie Tekstów pod SEO
AI może analizować istniejące treści na stronie internetowej i sugerować modyfikacje, które poprawią jej optymalizację pod kątem wyszukiwarek (SEO). Może też generować nowe, zoptymalizowane treści.

### 9. Tworzenie Personalizowanych E-maili
Model AI może automatycznie generować i personalizować e-maile na podstawie danych o odbiorcach. Może to być przydatne w marketingu, obsłudze klienta lub komunikacji wewnętrznej w firmie.

### 10. Automatyczna Redukcja Treści do Kluczowych Punktów
Model AI może przekształcać długie teksty w listy z kluczowymi punktami lub wypunktowaniami, co ułatwia szybkie przeglądanie najważniejszych informacji bez konieczności czytania całości.
